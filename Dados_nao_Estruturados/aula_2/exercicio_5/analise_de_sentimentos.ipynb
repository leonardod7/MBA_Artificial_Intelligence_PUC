Para um documento

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Frase de exemplo em inglês
sentence = "eu odeio pizza"

# Criar um objeto SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Analisar a sentença
scores = analyzer.polarity_scores(sentence)

# Imprimir os resultados
print(scores)
```

Para varios documentos

```python
import nltk
nltk.download('vader_lexicon')
from nltk.sentiment.vader import SentimentIntensityAnalyzer

# Lista de frases de exemplo em inglês
sentences = ["I love pizza!", "I hate waking up early.", "The sun is shining today."]

# Criar um objeto SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Loop através de cada sentença
for sentence in sentences:
    # Analisar a sentença
    scores = analyzer.polarity_scores(sentence)

    # Imprimir os resultados
    print(sentence)
    print(scores)
    print()

```
