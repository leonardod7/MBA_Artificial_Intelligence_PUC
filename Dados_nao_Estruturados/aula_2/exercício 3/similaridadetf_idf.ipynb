```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# define as duas strings para comparar
string1 = "esta é uma frase de exemplo"
string2 = "Esta é outra frase de exemplo"
```

```python
# cria um vetorizador TF-IDF
vectorizer = TfidfVectorizer()
```

```python
# transforma as strings em vetores TF-IDF
vector1 = vectorizer.fit_transform([string1])
vector2 = vectorizer.transform([string2])
```

```python
# calcula a similaridade do cosseno entre os vetores TF-IDF
similarity = cosine_similarity(vector1, vector2)
```

```python
# imprime o resultado
print(similarity)
#Neste exemplo, criamos um vetorizador TF-IDF usando a classe TfidfVectorizer da biblioteca scikit-learn. Em seguida, usamos o método fit_transform para transformar a primeira string em um vetor TF-IDF e o método transform para transformar a segunda string em um vetor TF-IDF. Finalmente, usamos a função cosine_similarity para calcular a similaridade do cosseno entre os dois vetores TF-IDF, o que nos dá um valor entre 0 e 1 que representa a similaridade entre as duas strings. O resultado é impresso no console.
```
