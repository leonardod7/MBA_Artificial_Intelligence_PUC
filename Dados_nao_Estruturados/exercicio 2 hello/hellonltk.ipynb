```
print(word_tokenize(noticia))
```

**Stop words**

```
nltk.download('stopwords')
from nltk.corpus import stopwords
```

```
stopWords_list = list(stopwords.words('portuguese'))
```

```
words = word_tokenize(noticia)
```

```
print(words)
```

```
print(stopWords_list)
```

```
clean_words = [word for word in words if not (word in stopWords_list)]
```

```
print(clean_words)
```

```
from collections import Counter

# Converte a lista em um dicionário com contagem de valores
word_counts_clean = Counter(clean_words)
a = word_counts_clean


```

```
word_counts_clean
```

```
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text = clean_words
wordcloud = WordCloud(width=1600, height=800, max_font_size=200).fit_words(a)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
