#Apresentação Pacotes

```python
# Instalar os pacotes
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Gerar dados artificialmente
x = list(range(1, 501))
y = np.random.normal(size=500)
```

```python
# plot de linhas
plt.figure(figsize=(20,5))  # aumenta o tamanho do plot
plt.plot(x, y)  # faz o plot de linhas

# INfos do gráficos
plt.title('Título do Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

plt.show()  # suprime o retorno da função (texto)
```

```python
# plot de linhas (mudar cor, marcador e estilo de linha)
plt.figure(figsize=(20,5))  # aumenta o tamanho do plot
plt.plot(x, y, 'g+-.')  # faz o plot de linhas (olhar fmt na documentação)

# INfos do gráficos
plt.title('Título do Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

plt.show()  # suprime o retorno da função (texto)
```

```python
# Gráfico de dispersão
plt.scatter(x, y)

# INfos do gráficos
plt.title('Título do Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')

plt.show()
```

```python
y_2 = np.random.normal(size=50)
x_2 = list(range(1,51))
```

```python
# Gráfico de dispersão (duas bases)
plt.scatter(x, y, label='base_1', color='orange')
plt.scatter(x_2, y_2, label='base2')

# INfos do gráficos
plt.title('Título do Gráfico')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.legend()  # coloca a legenda baseada nos labels do scatter

plt.show()
```

```python
# Boxplot
fig = plt.boxplot(y, labels=['Variável'], showmeans=True)

# Infos do gráfico
plt.title('Meu Boxplot')

plt.show()
```

```python
# retorna outliers do primeiro (índice 0) boxplot do gráfico fig
fig['fliers'][0].get_ydata()
```

```python
# 2 Boxplots em um mesmo gráfico
fig = plt.boxplot([y, y_2], labels=['Variável1', 'Variável2'], showmeans=True)

# Infos do gráfico
plt.title('Meu Boxplot')

plt.show()
```

```python
fig['fliers']
```

```python
# Histograma
plt.hist(y, bins=20)

# INfos do gráficos
plt.title('Meu Histograma')
plt.xlabel('Eixo x')
plt.ylabel('Frequência')

plt.show()
```

```python
# 2 Histogramas
plt.hist(y_2, bins=20, label='base1')
plt.hist(y, bins=20, alpha=0.5, label='base2')  # alpha é transparência

# INfos do gráficos
plt.title('Meu Histograma')
plt.xlabel('Eixo x')
plt.ylabel('Frequência')
plt.legend()

plt.show()
```
