### Tipo Sequencia

```
# Strings
d = 'python PUC'
type(d)
```

```
# acesso elementos da minha sequencia
d[3]
```

```
# de trás para frente: últimmo elemento
d[-1]
```

```
# Listas
lista1 = []
type(lista1)
```

```
# lista com diferentes tipos de dados
lista1 = ['puc', 'python', 1, 3.6, True, [1,2,3,4]]
lista1
```

```
# adicionar dados a lista
lista1.append('manoela')
lista1
```

```
# add em uma posição específica
lista1.insert(3, 567)
lista1
```

```
# Acessar elemento de uma lista
lista1[0]
```

```
lista1[6]
```

```
lista1[-2]
```

```
# remover elementos de uma lista
lista1.remove(1) #passar o elemento a ser removido
lista1
```

```
lista1.pop(2) # índice do elemento a ser removido
lista1
```

```
# Tuplas
tupla1 = ()
type(tupla1)
```

```
tupla1 = ('puc', 'cce', 1, 2222)
tupla1
```

```
# acessar elementos em uma tupla
tupla1[0]
```

```
tupla1[-2]
```

### Tipo Set

```
set1 = set()
type(set1)
```

```
set1 = set(['puc', 'cce', 'python', 'rio', 'puc', 1, 3.5, q2])
set1
```

```
# adicionar elementos
set1.add(8)
set1
```

```
set1.add('puc')
set1
```

```
set1.update([518,'manoela', 3, 1, 23]) # adiciona vários elementos ao set
set1
```

```
# acessar elemento (obs: não mantem ordenação)
print(list(set1))
list(set1)[7]
```

```
set1.remove(1)
set1
```
