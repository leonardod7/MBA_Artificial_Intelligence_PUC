#Exercício 1: crie uma função que receba um número como parâmetro e calcule o fatorial do mesmo

```
def fatorial(num):
  fat = 1
  for i in range(1, num + 1):
    fat *= i
  return fat
```

```
fatorial(5)
```

#Exercício 2: Crie uma função que receba o valor do raio de um círculo e retorne o valor da área do mesmo. Utilize o pacote math.


```
import math
def area_circulo(raio):
  return math.pi * raio**2
```

```
area_circulo(2)
```

Altere a função criada anteriormente para que ela retorne, além do valor da área do círculo, retorne também o diâmetro e o perímetro.


```
def area_circulo(raio):
  area = math.pi * raio**2
  diametro = 2*raio
  perimetro = 2*math.pi*raio
  return area, diametro, perimetro
```

```
area_circulo(4)
```

#Exercício 3: Função para LGN


```
import numpy as np
def lgn(exp):
  cont = 0
  for i in range(exp):
    x = np.random.normal()
    if(x > -1 and x < 1):
        cont += 1
  return cont / exp
```

```
print(lgn(10))
print(lgn(100))
print(lgn(1000))
print(lgn(10000))
print(lgn(100000))
```

Após validação da função, inclua um parâmetro opcional de intervalo a ser validado. O valor default deve ser -1 e 1.


```
def lgn(exp, intervalo_inferior=-1, intervalo_superior=1):
  cont = 0
  for i in range(exp):
    x = np.random.normal()
    if(x > intervalo_inferior and x < intervalo_superior):
        cont += 1
  return cont / exp
```

```
lgn(100000)
```

```
lgn(100000, 1, 2)
```

```
lgn(1000000, 2, 3)
```
