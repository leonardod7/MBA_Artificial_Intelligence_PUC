Criar array fornecendo os valores

```
x = np.array([[1,2,3,4],[5,6,7,8]])
x
```

```
x.shape
```

```
x.size #num de elementos  
```

```
x.ndim #numero de dimensoes
```

```
x.dtype #tipo dos dados
```

```
x.sum() #soma de tds os elementos
```

```
x.max() #maior valor
```

```
x.min() #menor valor
```

acessar elementos dentro de array

```
# acessar o primeriro elemento
x[0,0]
```

```
# elemento da segunda linha e segunda coluna
x[1,1]
```

```
# multiplos elementos
x[0, 1:3] # inclusive:exclusive. entoa nao pega col de idx 3
```

##Operações com arrays

```
a = np.array([[1,2], [4,5]])
b = np.array([[10,20], [30,40]])

print(a)
print(b)
```

```
# Multiplicação (element-wise)
2 * a
```

```
2 + a
```

```
# Matricial
a.dot(b)
```
