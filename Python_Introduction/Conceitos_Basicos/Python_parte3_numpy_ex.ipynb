# Exercício numpy
comparação operação vetorizada vs. não vetorizada 

```
import numpy as np
from time import process_time
```

```
#crio minha lista com 1000k elementos
lista = [1] * 1000000
len(lista)
```

```
# fazer operação nao vetorizada
tic = process_time()
dot = 0

for i in range(len(lista)):
  dot += lista[i] * lista[i]

toc = process_time()

print('Operação não vetorizada = ' + str(dot))
print('Tempo decorrido: ' + str(toc-tic) + 's')
```

```
# fazer operacao vetorizada
a = np.asarray(lista)

tic = process_time()
dot = a.dot(a)
toc = process_time()

print('Operação vetorizada = ' + str(dot))
print('Tempo decorrido: ' + str(toc-tic) + 's')
```
