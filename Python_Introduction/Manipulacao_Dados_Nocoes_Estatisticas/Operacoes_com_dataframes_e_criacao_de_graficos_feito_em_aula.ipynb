# Manipulação de DataFrame

Podemos criar uma nova coluna informando quem ganhou a partida. Para isso, podemos adotar a seguinte notação: 

- Empate: recebe o número 0
- Home_team: recebe o número 1
- Away_team: recebe o número 2

Essa informação de quem ganhou irá virar uma nova coluna de *df* chamada *rs*.

```python
df = df.dropna()
df
```

## Fazendo uso de função convencional

```python
df.columns
```

```python
A = df['home_score']-df['away_score']
A
```

```python
df.info()
```

```python
def funcao(s):
  output = []
  for i in s:
    if i == 0: # empate
      output.append(0)
    elif i > 0: # time da casa ganhador
      output.append(1)
    else: # time convidado ganhador
      output.append(2)
  return output 
```

```python
df['rs'] = funcao(A)
```

```python
df.head(1)
```

## Fazendo uso de função anônima 

- compreensão de listas: 

```python
funcao2 = lambda A: [0 if x == 0 else 1 if x > 0 else 2 for x in A]
```

```python
df['rs2'] = funcao2(A)
```
