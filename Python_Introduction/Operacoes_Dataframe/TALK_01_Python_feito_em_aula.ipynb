# Atividade 1

Desenvolva uma função que receba dois números como variáveis de entrada. Essa função irá retornar a informação de que o primeiro número é ou não é múltiplo do segundo número através de: 
- True 
- False


Exemplo:

```
Verifica(10,2)
>>  TRUE
```

 



````python
def Verifica(n1,n2):
  if n1%n2 == 0:
    return True
  else: # n1%n2 != 0
    return False 
````

````python
Verifica(10,2)
````

````python
Verifica(13,4)
````

# Atividade 5

Faça uma função que recebe dois números inteiros e retorna uma lista com os números inteiros que estão no intervalo compreendido por eles.

````python
'''
intervalo(10,2)
>> 3,4,5,6,7,8,9
'''

def intervalo(a,b):
  if a == b:
    return []
  elif a<b:
    res = []
    i = a+1
    while i < b:
      res.append(i)
      i = i + 1 # i += 1
    return res
  else: # a>b
    res = []
    i = b+1
    while i < a:
      res.append(i)
      i = i + 1
    return res 
````

````python
intervalo(40,34)
````

# Atividade 7

Crie uma função que recebe como variável de entrada uma lista e retorna outra lista contendo somente elementos do tipo *string*. 

````python
def funcao_str(a):
  out = []
  for i in a:
    if type(i) == str:
      out.append(i)
  return out 
````

````python
lista = ['a',3,65.-9,'bb','oi',34,True]
````

````python
funcao_str(lista)
````
